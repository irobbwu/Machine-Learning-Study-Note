# matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

用plt.show()来展示输入

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
```

颜色可以指定

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red')
```

样式也可以指定

`':'` 表示··········

`'-.'` 表示-·-·-·-·-·

`'--'` 表示----------

`'-'` 表示实线为默认

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
```

坐标轴的范围也可设置

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.axis([-1, 11, -2, 2])
```

给横纵轴起名

`plt.legend()`加入变量中label标明的图示

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.plot(x, siny, label = 'sin(x)')
plt.plot(x, cosy, color = 'red', linestyle = '--', label = 'cos(x)')
plt.legend()
plt.show()
```

`plt.title`加入标题

```python
plt.plot(x, siny, label = 'sin(x)')
plt.plot(x, cosy, color = 'red', linestyle = '--', label = 'cos(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Welcome to machine learning!')
plt.show()
```

## Scatter Plot 散点图

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color = 'red', linestyle = '--')
plt.show()
```

```python
x = np.random.normal(0, 1, 100000)
y = np.random.normal(0, 1, 100000)

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 100000)
y = np.random.normal(0, 1, 100000)

plt.scatter(x, y, alpha = 0.1)
# alpha 代表透明度
plt.show()
```
